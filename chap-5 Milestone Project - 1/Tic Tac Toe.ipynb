```python
def win_check(board,mark):
    patterns = (board[1:4], board[4:7], board[7:10], # rows
                board[1:9:3],board[2:9:3],board[3:10:3], # Columns
                board[1:10:4], board[3:9:2]) # across
    for pattern in patterns:
        if pattern == [mark]*3:
            return True
        
    return False
```

```python
import random

def choose_first():
    value = random.randint(1,2)
    print(f'player - {value} goes first')
    return 'player - '+str(value)
```

```python
def space_check(board,position):
    return board[position]==' '
```

```python
def full_board_check(board):
    return ' ' not in board
```

```python
def replay():
    
    answer = ''
    
    
    while answer not in ('Y','N'):
        
        answer = input("\nDo you want to play again. If yes, please select Y or N : ")
        
        if(answer in ('Y','N')):
            return True if answer=='Y' else False
        else:
            print("\nIncorrect marker, please select in 'Y' and 'N'")
    
    return False
```
